# Análise — NYC TLC Yellow Taxi

Este notebook responde às duas perguntas analíticas do case usando a camada Silver.

```python
SILVER_TABLE = "workspace.ifood_case.ny_tlc_silver"

if not spark.catalog.tableExists(SILVER_TABLE):
    raise RuntimeError(
        f"Tabela {SILVER_TABLE} não encontrada. Execute as etapas Bronze e Silver primeiro."
    )
```

## 1. Média do valor total por mês

Considera apenas corridas de táxis amarelos e ordena o resultado cronologicamente.

```python
avg_total_amount_by_month = spark.sql(f"""
    SELECT
        PickupYear,
        PickupMonth,
        ROUND(AVG(TotalAmount), 2) AS AvgTotalAmount
    FROM {SILVER_TABLE}
    WHERE TaxiCategory = 'YELLOW'
    GROUP BY PickupYear, PickupMonth
    ORDER BY PickupYear, PickupMonth
""")

display(avg_total_amount_by_month)
```

## 2. Média de passageiros por hora em maio de 2023

Agrupa as corridas de táxis amarelos pela hora de início da viagem.

```python
avg_passenger_count_by_hour = spark.sql(f"""
    SELECT
        PickupHour,
        ROUND(AVG(PassengerCount), 2) AS AvgPassengerCount
    FROM {SILVER_TABLE}
    WHERE TaxiCategory = 'YELLOW'
      AND PickupYear = 2023
      AND PickupMonth = 5
    GROUP BY PickupHour
    ORDER BY PickupHour
""")

display(avg_passenger_count_by_hour)
```
